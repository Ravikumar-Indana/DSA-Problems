```python
movies = pd.read_csv("movies.csv", index_col="Title")  # Reload the DataFrame
#movies["Gross"].str.replace('$', '', regex=False).str.replace(',', '', regex=False)  # Convert "Gross" column to float after removing '$' and ','
```

```python
movies["Gross"] = (
    movies["Gross"]
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
) # convert "Gross" column to float after cleaning the data
```
